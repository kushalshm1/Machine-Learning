# Regression for Prediction - just enter your data and price label 

```python
import numpy as np
import pandas as pd
import random
```

## Enter Data CSV file name here

```python
dataName = "Income_training.csv"
priceLabel = 'compositeHourlyWages'
```

```python
df = pd.DataFrame(pd.read_csv(dataName))
df = df/df.max()
df
```

```python
price = df[priceLabel]
features = df
features = features.drop([priceLabel],axis = 1)
NoOfFeatures = (len(features.T))
```

### Function for Mean Square Error

```python
def cost(weights,features,price):
    return np.mean(np.square(np.dot(features,weights) - price))
   
```

### Function for Gradient

```python
def gradients(weights,features,price):
    return np.mean(2*(np.dot(features,weights) - price))*weights.T
```

### Train Function

```python
def train(iters,lr):
    global features,price
    initialWeights = np.array(np.random.randn(1,NoOfFeatures))
    initialWeights = initialWeights[0]
    initialWeights = np.array(initialWeights)
    for i in range(0,iters):
        newWeights = initialWeights - lr*gradients(initialWeights,features,price)
        initialWeights = newWeights
        
    print("Your Machine is Trained")
    print("Error: ",cost(initialWeights,features,price))
    print("Trained Weights:")
    
    return initialWeights
    
```

```python
train(1000,0.000001)
```
